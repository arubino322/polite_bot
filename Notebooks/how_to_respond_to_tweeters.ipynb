```python
def positive_responses(tweet):

    response = ""
    
    thanks = ['thank', 'thanks']
    if any(x in tweet for x in thanks):
        response += "Thanks for being polite! "

    good_day = ['good day', 'wonderful day', 'great day', 'fabulous day', 'awesome day', 'safe day']
    if any(x in tweet for x in good_day):
        response += "You have a good day yourself! "
    
    appreciate = ['appreciate']
    if any(x in tweet for x in appreciate):
        response += "Appreciate your kindness! "
        
    if response != "":
        return response
    else:
        pass
```

```python
# some ideas of functions that can be applied to labeled tweets. what if it has both?
def negative_responses(tweet):
    
    response = ""

    wtf = ['wtf', 'what the fuck']
    if any(x in tweet.lower() for x in wtf):
        response += "Wtf?! "

    fuck_you = ['fuck you', 'you fuck', 'fucking asshole', 
                'fuck off', 'fuckin ass']
    if any(x in tweet.lower() for x in fuck_you):
        response += "You're not being very nice. How about we try being nice for a change? "

    curse = [' ass', 'asshole', 'retard', 'bitch', 'idiot', 'dumbass', 
             'jackass', ' shit', 'shitty']
    if any(x in tweet.lower() for x in curse):
        response += "Do you kiss your mother with that mouth? "

    wait = ['waiting', 'wait for', 'where is']
    if any(x in tweet.lower() for x in wait):
        response += "Try downloading the SubwayTime app for more info. "

    if response != "":
        return response
    else:
        pass
```

# We can improve on these by a whole lot. But let's move forward and take the next step, which is decide on how we want to reply to each tweet.

```python
predict_df = pd.read_csv('../data/predict_df_oct6.csv')
print(predict_df.shape )
predict_df.head() 
```

```python
def how_should_we_respond(row):
    if row['compound_score'] >= 0.75 and row['do_they_agree'] == 'all_positive':
        return positive_responses(row['tweet_text'])
    elif row['compound_score'] <= 0.75 and row['do_they_agree'] == 'all_negative':
        return negative_responses(row['tweet_text'])
    
    

predict_df['response'] = predict_df.apply(how_should_we_respond, axis=1)
    
```

```python
# how many responses did we get?
len(predict_df.loc[predict_df.response.notnull()])
```
